```python
import pdfplumber,re,pandas as pd, os,re, colorama
from colorama import Fore, Back, Style
colorama.init(autoreset=True)

def load_pdf(path, dont_find_table = True):
    pdf = pdfplumber.open(path)
    print(pdf.metadata)
    print('Total Pages: ', len(pdf.pages))
    if dont_find_table:
        return pdf
    for i, page in enumerate(pdf.pages):
        tb_found = page.find_tables()
        if len(tb_found) > 0:
            print('Page Index', i, 'Has ', len(tb_found), ' Table.')
    return pdf

def extract_table(pages):
    merged_tb = []
    for page in pages:
        tb = page.extract_table(
            table_settings={"horizontal_strategy": "lines"})
        if tb is None:
            continue
        merged_tb = merged_tb + tb
    return merged_tb

def save_to_xlsx(table, path):
    table_data = pd.DataFrame(table[1:], columns = table[0])
    table_data.to_excel(path, index = False)
    
home = './ICPC/'
output = './extracted/ICPC/'
file = '47ICPC济南.pdf'
path = home + file
output_path = output + file.replace('.pdf', '.xlsx')

pdf = load_pdf(path, False)
```

```python
table = extract_table(pdf.pages[18:46])
table
```

```python
from IPython.display import Markdown, display
refined = []
title = ['学校名称', '队伍名称', '队员1姓名', '队员2姓名', '队员3姓名', '队伍类型']
cn_team_index = 1
column_mapping = {
    6: 3,
    1: 0,
    7: 4,
    2: 1,
    9: 4,
    4: 1,
    5:2,
    3:2
}
cn_col = [0, 1, 2, 3, 4, 5]
en_col = []
row_skip = [[]]
row_replace = {

}
school_fix = {}
known_fixes = {}
refined.append(title)
print(title)


def printmd(string, list = None, color = 'red'):
    colorstr = "<span style='color:{}'>{}{}</span>".format(color, string, '' if list is None else list)
    display(Markdown(colorstr))


def try_fix(data, add):
    ori_add = add.copy()
    ori_data = data.copy()
    global title
    global column_mapping
    global known_fixes
    has_new_fix = False
    for mapper in column_mapping:
        if len(add) > mapper and add[mapper] is not None and len(add[mapper]) > 0:
            to = column_mapping[mapper]
            if add[mapper] not in known_fixes:
                known_fixes[add[mapper]] = data[to]
                has_new_fix = True
            data[to] = data[to] + add[mapper]
            add[mapper] = None
    if has_new_fix:
        printmd('--- Fixing Starts ---')
        printmd('Data To Fix: ', ori_data)
        printmd('Data To Add: ', ori_add)
        printmd('Fixed Data: ', data)
        printmd('--- Fixing Ends ---')
    else:
        printmd('--- Known Fixing ---', color='green')
    fix_succeed = True
    for i, item in enumerate(add):
        if item is not None and len(item) > 0:
            fix_succeed = False
            print('Cannot Fix Data: ',i, item)
    if not fix_succeed:
        raise Exception('Unprocessed Data: ', add)
    return data
    

for raw1 in table[2:]:
    raw = raw1.copy()
    baked = []
    baked2 = []
    for data in raw:
        if data is not None:
            baked.append(data)
            if len(data) > 0:
                baked2.append(data)
    #It's likely there are some empty strings
    if len(baked2) == len(title) and len(baked) > len(title):
        baked = baked2
    if baked in row_skip:
        continue

    if baked[0] in row_replace:
        baked = row_replace[baked[0]].copy()
    
    if len(baked) != len(title):
        refined[-1] = try_fix(refined[-1].copy(), raw)
        continue

    for i in cn_col:
        baked[i] = baked[i].replace('\n', '')
    for i in en_col:
        baked[i] = baked[i].replace('\n', ' ')
    
    if baked == title:
        continue

    if baked[1] in school_fix and int(baked[0]) in school_fix[baked[1]][1]:
        baked[1] = school_fix[baked[1]][0]
    #print(baked)
    # if baked[cn_team_index][0] in ['☆', '\uf0f3']:
    #     baked[cn_team_index] = baked[cn_team_index][1:]
    #     baked.append('Shadow')
    # else:
    #     baked.append('Primary')
    refined.append(baked)
refined
```
